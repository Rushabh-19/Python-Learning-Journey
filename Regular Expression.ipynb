Before you can use regular expressions in your program, you must import the library using "import re".
- You can use re.search() to see if a string matches a regular expression, similar to using the find() method for strings
- You can use re.findall() to extract portions of a string that match your regular expression, similar to a combination of find() and slicing: var[5:10]


## Using re.search() like find()

Following is without use of RegExp

```python
hand = open ('mbox-short.txt')
for line in hand:
    line = line. rstrip()
    if line. find('From:') >= 0 :
        print (line)
```

Following is with using RegExp

```python
import re
hand = open('mbox-short.txt')
for line in hand:
    line = line. rstrip ()
    if re. search('From:', line) :
        print (line)
```

## Using re.search() like startswith()

Following is without regExp

```python
hand = open ('mbox-short.txt')
for line in hand:
    line = line. rstrip ()
    if line. startswith ('From: '):
        print (line)
```

Following is with regExp

```python
import re
hand = open ('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    if re.search('^From:', line) :
        print (line)
```

```python
import re

x = 'From: Using the : character'
y = re.findall('^F.+:', x)

print (y)
```

```python
import re

x = 'From: Using the : character'
y = re.findall('^F.+?:', x)

print (y)
```

```python
import re
x = 'From stephen.marquard@uct.ac.za Sat Jan  5 09:14:16 2008'
y = re.findall('\S+?@\S+',x)
print(y)
```

```python
import re
fh = open('regex_sum_1687014.txt')

sum = 0
for lines in fh:
    n = re.findall('[0-9]+',lines)
    if len(n) == 0: continue
    for num in n:
        sum += int(num)
print (sum)
```
