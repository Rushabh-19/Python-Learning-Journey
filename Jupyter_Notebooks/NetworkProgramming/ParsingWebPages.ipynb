```python
import urllib.request, urllib.parse , urllib.error
from bs4 import BeautifulSoup

url = input ('Enter - ') #Ask for URL
html = urllib.request.urlopen(url).read() #open the URL and Read it. read() - will return whole block in a string.

soup = BeautifulSoup(html , 'html.parser') # Asking BeautifulSoup to parse the html for us
print ('---------------\n{}-----------------'.format(soup))
# Retriving all of the Anchor Tags
tags = soup ('a')  # this give a list ([.... , ... , ...]) of Anchor Tag in the Block
for tag in tags:
    print (tag.get('href' , None)) # data is in the form of key - Value pair i.e. href = 'http:// ____'. So here we have extracted only the Value for the key 'href'
```
